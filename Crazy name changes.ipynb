```python
import json
import Levenshtein
```

```python
d = json.load(open('export.json'))
```

```python
url_template = "https://yournextmp.com/person/{}"

first = True

for check_dist in range(5,1,-1):
    if first:
        print "## Levenshtein distance >= {}".format(check_dist)
    else:
        print "## Levenshtein distance = {}".format(check_dist)
    
    for p in d['persons']:
        last_name = None
        url = url_template.format(p['id'])

        for v in reversed(p['versions']):
            if last_name:
                dist = Levenshtein.distance(v['data']['name'], last_name)

                if (first and dist >= check_dist) or (dist == check_dist):
                    print u"[{3}]({0}): {1} -> {2}\n".format(url, last_name, v['data']['name'], p['name'])

            last_name = v['data']['name']
            
    if first:
        first = False
```
