```python
import json
from collections import Counter
d = json.load(open('data/export.json'))
```

### Helper functions

```python
def m_s(xs, k):
    """
        Turn a list of dictionaries into a dictionary given a key
    """
    return {x[k]: x for x in xs}
```

```python
def g(x, key):
    """
        Get the property of a Maybe dictionary
    """
    if x is None:
        return None
    else:
        return x.get(key)
```

```python
def l(x):
    if x is None:
        return None
    else:
        return x.lower()
```

## Look for duplicated email addresses

```python
emails = Counter()

for p in d['persons']:
    email = p.get('email', None)
    
    if email is not None:
        emails[email] += 1

for p in sorted(d['persons'], key=lambda x: x.get('email')):
    email = p.get('email', None)
 
    if email in emails and emails[email] > 1:
        print email, p['name'], "https://yournextmp.com/person/{}".format(p['id']) 
```

## Look for duplicate links

```python
def find_duplicate_link(link_name):
    link_counts = Counter()

    for p in d['persons']:
        p['links_d'] = m_s(p['links'], 'note')
        link = l(g(g(p['links_d'], link_name), 'url'))

        if link is not None:
            link_counts[link] += 1

    for p in sorted(d['persons'], key=lambda x: l(g(g(x['links_d'], link_name),'url'))):
        link = l(g(g(p['links_d'], link_name), 'url'))

        if link in link_counts and link_counts[link] > 1:
            print p['links_d'][link_name]['url'], p['name'], "https://yournextmp.com/person/{}".format(p['id']) 
```

### Duplicate home pages

```python
find_duplicate_link('homepage')
```

### Duplicate Facebook page

```python
find_duplicate_link('facebook page')
```

### Duplicate Facebook profile

```python
find_duplicate_link('facebook personal')
```

### Duplicate PPC page

```python
find_duplicate_link('party PPC page')
```
